### 2.- Matplotlib

##### Gráficas en 2D

```python
import matplotlib.pyplot as plt
```

```python
# X
x = np.linspace(0, 2*np.pi, 200)

# Y
y1 = np.sin(x)
y2 = -2.0*np.cos(x)
y3 = np.tan(x)
```

```python
# Graficar datos
plt.figure(2)
plt.plot(x,y1,"r", label = "sin(x)")
plt.plot(x,y2,"b", label = "-2cos(x)")
plt.plot(x,y3,"g", label = "tan(x)")

# Limitar rango
plt.xlim([0,6])
plt.ylim([-10,10])

# Etiquetas
plt.title("Funciónes trigonométricas")
plt.xlabel("Tiempo (s)")
plt.ylabel("Señal")
plt.legend(loc = "lower left")

# Mostrar gráficas
plt.grid()
plt.show()
```

##### Gráficas tipo Scatter

```python
x = np.zeros([400,2])
y = np.ones(400)

# Crear clusters
A = np.zeros([200,2])
B = np.zeros([200,2])

# Calcular clusters
A[:,0] = 0.5 + 0.2*np.random.rand(200) - 0.2*np.random.rand(200)
A[:,1] = 0.8 + 0.2*np.random.rand(200) - 0.2*np.random.rand(200)
y1 = np.ones(200)

B[:,0] = 0.3 + 0.2*np.random.rand(200) - 0.2*np.random.rand(200)
B[:,1] = 0.7 + 0.2*np.random.rand(200) - 0.2*np.random.rand(200)
y2 = 2*np.ones(200)

# Adjuntar arreglos
x[0:200,:] = A
x[200:400,:] = B
y[0:200] = y1
y[200:400] = y2   

# Crear gráfica
plt.scatter(x[:,0],x[:,1], c=y, s=40, cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

```python
# Crear clusters
C = np.zeros([200,2])
D = np.zeros([200,2])

# Calcular clusters
C[:,0] = 0.4 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
C[:,1] = 0.4 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)

D[:,0] = 0.6 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
D[:,1] = 0.6 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)

# Adjuntar arreglos
x[0:200,:] = C
x[200:400,:] = D
y[0:200] = y1
y[200:400] = y2   

# Crear gráfica
plt.scatter(x[:,0],x[:,1], c=y, s=40, cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

```python
from mpl_toolkits import mplot3d
%matplotlib notebook

# Crear gradilla uniforme
x = np.zeros([400,2])

#Asignar valores
for i in range(20):
    
    # X fija
    x[20*i:20*(i+1),0] = i/20.0
    x[20*i:20*(i+1),1] = np.linspace(0,1,20)
    
# Asignar valor z
z = np.power(x[:,0]-0.5,3) + np.power(x[:,1]-0.5,3)

# Crear gráfica
ax = plt.axes(projection="3d")
ax.scatter(x[:,0],x[:,1], z, c=z, marker="o", cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

### 3.- Programación Orientada a Objetos

##### Herencias en Python

```python
class personal:
    # Constructor
    def __init__(self,nombre,edad,curp):
        
        #Definir atributos
        self.nombre = nombre
        self.edad = edad
        self.curp = curp
        
    # Métodos de instancia
    def modNombre(self,n_nombre):
        self.nombre = n_nombre
    
    def modEdad(self,n_edad):
        self.edad = n_edad
        
    def modCurp(self,n_curp):
        self.curp = n_curp
```

```python
class alumno(personal):
    # Constructor
    def __init__(self,nombre,edad,curp,carrera,semestre):
        
        # Llamar a constructor de la clase padre
        personal.__init__(self,nombre,edad,curp)
        
        # Definir atributos
        self.carrera = carrera
        self.semestre = semestre
        
    # Agregar métodos adicionales
    def modCarrera(self,n_carrera):
        self.carrera = n_carrera
    
    def modSemestre(self,n_semestre):
        self.semestre = n_semestre

class profesor(personal):
    # Constructor
    def __init__(self,nombre,edad,curp,departamento):
        
        # Llamar a constructor de la clase padre
        personal.__init__(self,nombre,edad,curp)
        
        # Definir atributos
        self.departamento = departamento
        
    # Agregar métodos adicionales
    def modDepartamento(self,n_departamento):
        self.departamento = n_departamento
```

```python
# Instancias
per =  personal("Gabriela Guerrero",30,"AAAA111111AAAAAA11")
alu = alumno("Javier Millas",18,"BBBB222222BBBBBB22","ISC",2)
pro = profesor("Alejando Torres",50,"CCCC333333CCCCCC33","Negocios") 
```

```python
# Atributos originales
print("Personal:""\n"
      "Nombre: " + per.nombre + "\n"
      "Edad: " + str(per.edad) + "\n"
      "CURP: " + per.curp + "\n"
      "-----------------------------")
print("Alumno:""\n"
      "Nombre: " + alu.nombre + "\n"
      "Edad: " + str(alu.edad) + "\n"
      "CURP: " + alu.curp + "\n"
      "Carrera: " + alu.carrera + "\n"
      "Semestre: " + str(alu.semestre) + "\n"
      "-----------------------------")
print("Profesor:""\n"
      "Nombre: " + pro.nombre + "\n"
      "Edad: " + str(pro.edad) + "\n"
      "CURP: " + pro.curp + "\n"
      "Departamento: " + pro.departamento + "\n")
```

```python
# Llamar métodos
per.modNombre("Susana Guerrero")
per.modEdad(32)
per.modCurp("DDDD444444DDDDDD44")

alu.modNombre("Mariana Flores")
alu.modEdad(21)
alu.modCurp("EEEE555555EEEEEE55")
alu.modCarrera("ARQ")
alu.modSemestre(7)

pro.modNombre("Valeria Ibarra")
pro.modEdad(43)
pro.modCurp("FFFF666666FFFFFF66")
pro.modDepartamento("Humanidades y Educación")
```
