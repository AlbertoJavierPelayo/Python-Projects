# Creaci칩n y manejo de gr치ficas en Python

# 1.- Importar librerias

matplotlib es una libreria de python que permite realizar diferentes tipos de gr치ficas de manera muy sencilla, incluyendo de dos y tres dimensiones

```python
# Instalar en terminal -->  pip3 install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# 2.- Gr치ficas en dos dimensiones

```python
# Vector X
x = np.linspace(0, 6.28, 200)

# Vector Y = Sin(x)
y = np.sin(x)

# Graficar datos
plt.figure(1)
plt.plot(x,y,"y")
plt.title("Funci칩n sine")
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
plt.grid()
plt.show()
```

```python
# Vector Y = Sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Graficar datos
plt.figure(2)
plt.plot(x,y,"y")
plt.plot(x,y2,"b")
plt.plot(x,y3,"r")

# Limitar rango
plt.xlim([0,6])
plt.ylim([-4,4])

# Etiquetas
plt.title("Funci칩nes")
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")

# Mostrar gr치ficas
plt.grid()
plt.show()
```

```python
# Multiples gr치ficas con tama침o definido
plt.figure(2, figsize = (10,10))

# Gr치fica Sin
plt.subplot(3,1,1)
plt.plot(x,y,"y")
plt.grid()

# Gr치fica Cos
plt.subplot(3,1,2)
plt.plot(x,y2,"b")
plt.grid()

# Gr치fica Sin
plt.subplot(3,1,3)
plt.plot(x,y3,"r")
plt.xlim([0,6])
plt.ylim([-4,4])
plt.grid()
```

# 3.- Graficas tipo scatter

Las gr치ficas de tipo scatter permiten representar puntos de dos dimensiones que poseen alg칰n tipo de clasificaci칩n

```python
# Crear serie de datos
x1 = np.zeros([200,2])

# Calcular cluster
x1[:,0] = 0.2 + 0.1*np.random.rand(200)
x1[:,1] = 0.2 + 0.1*np.random.rand(200)

# Eje de color
y1 = 0*np.ones(200)

# Crear gr치fica
plt.figure(4)
plt.scatter(x1[:,0],x1[:,1], c = y1, s = 40, cmap = plt.cm.Spectral)
plt.grid()
plt.show()
```

```python
# Crear clusters
x2 = np.zeros([200,2])
x3 = np.zeros([200,2])

# Calcular clusters
x2[:,0] = 0.6 + 0.2*np.random.rand(200)
x2[:,1] = 0.6 + 0.2*np.random.rand(200)
y2 = np.ones(200)

x3[:,0] = -0.1 + 0.2*np.random.rand(200)
x3[:,1] =  0.6 + 0.2*np.random.rand(200)
y3 = 2*np.ones(200)

# Crear gr치fica
plt.figure(5)
plt.scatter(x1[:,0],x1[:,1], c=y1, s=40, cmap=plt.cm.Spectral)
plt.scatter(x2[:,0],x2[:,1], c=y2, s=40, cmap=plt.cm.Spectral)
plt.scatter(x3[:,0],x3[:,1], c=y3, s=40, cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

```python
# Juntar arreglos en uno solo
x = np.zeros([600,2])
y = np.ones(600)
             
# Adjuntar arreglos
x[0:200,:] = x1
x[200:400,:] = x2
x[400:,:] = x3
y[0:200] = y1
y[200:400] = y2
y[400:] = y3      

# Crear gr치fica
plt.figure(5)
plt.scatter(x[:,0],x[:,1], c=y, s=40, cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```

La clase mpl_toolkits provee herramientas para representar puntos tridimensionales de manera sencilla

```python
from mpl_toolkits import mplot3d
%matplotlib inline
#%matplotlib inline

# Crear gradilla uniforme
x = np.zeros([400,2])

#Asignar valores
for i in range(20):
    
    # X fija
    x[20*i:20*(i+1),0] = i/20.0
    x[20*i:20*(i+1),1] = np.linspace(0,1,20)
    
# Asignar valor z
z = np.power(x[:,0],2) + np.power(x[:,1],2)

# Crear gr치fica
plt.figure(6)
ax = plt.axes(projection="3d")
ax.scatter(x[:,0],x[:,1], z, c=z, marker="o", cmap=plt.cm.Spectral)
plt.grid()
plt.show()
```
